```python
start_sym = grammar.start()
```

```python
new_start = Nonterminal("NEW_START")
rules.append(Production(new_start, [start_sym]))
```

```python
def handle_long_rules(rules):
    new_rules = list()
    for j in range(len(rules)):
        rhs = list(rules[j].rhs())
        lhs = rules[j].lhs()
        if len(rhs) > 1:
            for i in range(len(rhs) - 1):
                if i == len(rhs) - 2:
                    new_prod = Production(lhs, rhs[i:])
                    new_rules.append(new_prod)
                else:
                    new_sym = Nonterminal(str(j) + str(i))
                    new_prod = Production(lhs, [rhs[i], new_sym])
                    lhs = new_sym
                    new_rules.append(new_prod)
        else:
            new_rules.append(rules[j])
    return new_rules
```

```python
def handle_unary_rules(rules):
    new_rules = copy.deepcopy(rules)
    
    for j in range(len(rules)):
        
        rhs = list(rules[j].rhs())
        lhs = rules[j].lhs()
        if len(rhs) == 1 and type(rhs[0]) == type(lhs):
            rplt = rhs
            rpwth = lhs
            tmp_rules = list()
            for i in range(len(rules)):
                if i != j:
                    lhs_tmp = rules[i].lhs()
                    rhs_tmp = rules[i].rhs()
                    if rplt == lhs_tmp:
                        lhs_tmp = rpwth
                    if rplt in rhs_tmp:
                        rhs_tmp = [rhs_tmp[i] if rhs_tmp[i] != rplt else rpwth]
                    new_rule = Production(lhs_tmp, rhs_tmp)
                    tmp_rules.append(new_rule)
            new_rules = copy.deepcopy(tmp_rules)
        else:
            new_rules.append(rules[j])
            
    return new_rules
            
```

```python
rules = handle_long_rules(rules)
```
